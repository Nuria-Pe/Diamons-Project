### 2. Cambio a variables numéricas

#### Primero con la columna "cut"

```python
data.cut.value_counts()
```

```python
height_map = {
    "Ideal": 4,
    "Premium": 3,
    "Very Good": 2,
    "Good": 1,
    "Fair": 0
}
```

```python
data["cut_name"] = data.cut.map(height_map)
```

```python
data.head()
```

#### Ahora columna "color"

```python
data.color.value_counts()
```

```python
height_num = {
    "G": 7,
    "E": 6,
    "F": 5,
    "H": 4,
    "D": 3,
    "I": 2,
    "J": 1
}
```

```python
data["color_name"] = data.color.map(height_num)
data
```

#### Por último la de "Clarity"

```python
data.clarity.value_counts()
```

```python
clarity_num = {
    "SI1": 7,
    "VS2": 6,
    "SI2": 5,
    "VS1": 4,
    "VVS2": 3,
    "VVS1": 2,
    "IF": 1,
    "I1": 0
}
```

```python
data["clarity_name"] = data.clarity.map(clarity_num)
data.head()
```

```python
data.drop(["cut", "clarity", "color"], axis=1, inplace=True)
```

```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
%matplotlib inline
%config Inlinebackend.figure_format = 'retina'

import seaborn as sns
sns.set_context('poster')
sns.set(rc={'figure.figsize': (16., 9.)})
sns.set_style('whitegrid')

# Modeling libraries
import statsmodels.formula.api as smf # welcome!!
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import learning_curve, GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, RandomForestRegressor
from sklearn.model_selection import cross_validate, cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn import set_config
set_config(display='diagram')

import pandas as pd
np.random.seed(123)

import warnings
warnings.filterwarnings('ignore')
```

```python

# Split target and predictors
X = data.drop("price", axis=1)
y = data["price"]


# split data in train/ test/ 
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.20, 
                                                    random_state=123)
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(X_train, y_train)
```

```python
rf.predict(X_test)[:10]
```

```python
mean_squared_error(
    y_true=y_test,
    y_pred=rf.predict(X_test)
)
```
